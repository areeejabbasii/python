```python
my_list = [1, 2] #append
my_list.append(3)
print(my_list)  # Output: [1, 2, 3]
```

```python
my_list = [1, 2] #extend
my_list.extend([3, 4])
print(my_list)  # Output: [1, 2, 3, 4]
```

```python
my_list = [1, 3]
my_list.insert(1, 2)  # Insert 2 at index 1
print(my_list)  # Output: [1, 2, 3]
```

```python
my_list = [1, 2, 3, 2] #remove
my_list.remove(2)
print(my_list)  # Output: [1, 3, 2]
```

```python
my_list = [1, 2, 3]
item = my_list.pop()  # Removes the last element
print(item)  # Output: 3
print(my_list)  # Output: [1, 2]
```

```python
my_list = [1, 2, 3] #index
print(my_list.index(2))  # Output: 1
```

```python
my_list = [1, 2, 2, 3] #count
print(my_list.count(2))  # Output: 2
```

```python
my_list = [1, 2, 3] #reverse
my_list.reverse()
print(my_list)  # Output: [3, 2, 1]
```

```python
my_list = [3, 1, 2]
my_list.sort() #sort
print(my_list)  # Output: [1, 2, 3]
```

```python
my_list = [1, 2, 3]
my_list.clear() #clear
print(my_list)  # Output: []
```

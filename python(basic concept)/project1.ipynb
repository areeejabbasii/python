rules for **identifier**



1. Allowed Characters
An identifier can only contain:
Letters (A-Z, a-z)
Digits (0-9)
Underscore (_)
2. Cannot Start with a Digit
Identifiers must not begin with a digit.
Example:
✅ name1
❌ 1name
3. Case Sensitivity
Python is case-sensitive.
Example:
Name and name are considered different identifiers.
4. Cannot Use Reserved Keywords
Reserved keywords in Python (e.g., if, else, True, False, None, etc.) cannot be used as identifiers.
5. Cannot Contain Special Characters
Identifiers cannot include symbols like @, #, $, %, &, etc.
6. No Spaces Allowed
Spaces are not allowed in identifiers. Use underscores (_) instead to separate words.
Example:
✅ user_name
❌ user name
7. Unlimited Length
Identifiers can be of any length but should be kept readable and meaningful.
8. Dunder Names
Identifiers that begin and end with double underscores (__) are reserved for special purposes in Python (e.g., __init__, __str__).
Examples:
Valid Identifiers:
x
my_variable
temperature_23
_hidden
Name
Invalid Identifiers:
3variable (starts with a digit)
user-name (contains a hyphen)
class (reserved keyword)
user name (contains space)

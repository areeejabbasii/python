```python
# prompt: write a code that  takes three input from user and print the number is positive negative or zero

# Previous code is not relevant to the task.

num1 = float(input("Enter first number: "))
if num1 > 0:
  print("First number is positive")
elif num1 < 0:
  print("First number is negative")
else:
  print("First number is zero")

num2 = float(input("Enter second number: "))
if num2 > 0:
  print("Second number is positive")
elif num2 < 0:
  print("Second number is negative")
else:
  print("Second number is zero")

num3 = float(input("Enter third number: "))
if num3 > 0:
  print("Third number is positive")
elif num3 < 0:
  print("Third number is negative")
else:
  print("Third number is zero")
```

```python
# prompt: write that takes input three number from user and print largest number

# take three numbers as input from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

# find the largest number using max() function
largest_number = max(num1, num2, num3)

# print the largest number
print("The largest number is:", largest_number)
```

```python
# prompt: write python code which asks 2 numbers from me return sum of these number

def sum_two_numbers():
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        result = num1 + num2
        print("The sum of", num1, "and", num2, "is:", result)
    except ValueError:
        print("Invalid input. Please enter numbers only.")

sum_two_numbers()
```

guess the number from computer


```python
# prompt: write a program of guessing a random number game by computer

import random

def guess(x):
    random_number = random.randint(1, x)
    guess = 0
    while guess != random_number:
        try:
            guess = int(input(f"Guess a number between 1 and {x}: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
        if guess < random_number:
            print("Too low.")
        elif guess > random_number:
            print("Too high.")
    print(f"Congratulations! You guessed the number {random_number} correctly!")

guess(10)
```

guess the num to input from user

```python
# prompt: write a code to guess a number by user

import random

def computer_guess(x):
    low = 1
    high = x
    feedback = ''
    while feedback != 'c':
        if low != high:
            guess = random.randint(low, high)
        else:
            guess = low  # could also be high since low = high
        feedback = input(f'Is {guess} too high (H), too low (L), or correct (C)?? ').lower()
        if feedback == 'h':
            high = guess - 1
        elif feedback == 'l':
            low = guess + 1
    print(f'Yay! The computer guessed your number, {guess}, correctly!')

computer_guess(100)
```

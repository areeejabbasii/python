```python
# prompt: create  a word guessing game use python package random  also use loop keep track of user tries when user guess a number show the number being bused by user

import random

# Generate a random number between 1 and 5
secret_number = random.randint(1, 5)

# Initialize the number of tries
tries = 0

# Start the game loop
while True:
    try:
        # Get the user's guess
        guess = int(input("Guess a number between 1 and 5: "))


        tries += 1


        if guess == secret_number:
            print(f"Congratulations! You guessed the number {secret_number} in {tries} tries.")
            break
        elif guess < secret_number:
            print("Too low! Try again.")
        else:
            print("Too high! Try again.")

    except ValueError:
        print("Invalid input. Please enter a number.")
```

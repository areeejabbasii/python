```python

# Collecting Student's Information
name = input("Enter your name: ")
father_name = input("Enter your father's name: ")
date = input("Enter today's date (DD/MM/YYYY): ")
roll_no = input("Enter your roll number: ")

# Displaying Student's Information
print(f"\nStudent Information:")
print(f"Name: {name}")
print(f"Father's Name: {father_name}")
print(f"Roll Number: {roll_no}")
print(f"Date: {date}\n")

# Initialize subject list and total marks
subjects = ["Math", "Physics", "Chemistry", "Computer Science", "English"]
total_marks_per_subject = 100
total_marks = len(subjects) * total_marks_per_subject
obtained_marks = 0  # Initialize total obtained marks

# Input marks for each subject
subject_results = []
for subject in subjects:
    while True:
        try:
            marks = float(input(f"Enter marks obtained in {subject} (0-100): "))
            if 0 <= marks <= 100:
                subject_results.append((subject, marks))
                obtained_marks += marks
                break
            else:
                print("Marks should be between 0 and 100. Try again.")
        except ValueError:
            print("Invalid input! Please enter a valid number.")

# Calculate percentage and grade
percentage = (obtained_marks / total_marks) * 100
if percentage >= 90:
    overall_grade = "A+"
elif percentage >= 80:
    overall_grade = "A"
elif percentage >= 70:
    overall_grade = "B"
elif percentage >= 60:
    overall_grade = "C"
elif percentage >= 50:
    overall_grade = "D"
else:
    overall_grade = "F"

# Display Result Card
print("\nResult Card:")
print("="*50)
print(f"Name: {name}")
print(f"Father's Name: {father_name}")
print(f"Roll Number: {roll_no}")
print(f"Date: {date}")
print("="*50)
print(f"{'Subject':<20}{'Marks Obtained':<15}{'Grade':<10}")
print("-"*50)

# Calculate and display subject-wise grades
for subject, marks in subject_results:
    if marks >= 90:
        grade = "A+"
    elif marks >= 80:
        grade = "A"
    elif marks >= 70:
        grade = "B"
    elif marks >= 60:
        grade = "C"
    elif marks >= 50:
        grade = "D"
    else:
        grade = "F"
    print(f"{subject:<20}{marks:<15}{grade:<10}")

print("="*50)
print(f"Total Marks: {total_marks}")
print(f"Marks Obtained: {obtained_marks}")
print(f"Percentage: {percentage:.2f}%")
print(f"Overall Grade: {overall_grade}")
print("="*50)

# Final Remark
if overall_grade == "F":
    print("Remark: Unfortunately, you have failed. Better luck next time!")
else:
    print("Remark: Congratulations on passing!")
```
